```r
data = read.table('./subdingwei.txt',header =T,sep = '\t')
```

```r
head(data,6)
```

```r
library(ggplot2)
```

```r
library(ggthemes)
ggplot(data,aes(x = reorder(type,HIST1H4C),y=HIST1H4C,
           fill = type,
           label = HIST1H4C))+
    geom_col(show.legend = FALSE) +
    coord_flip() +
    geom_text(nudge_x = 0.1) +
    xlab('') +
    ggtitle('Expression level of significant HIST1H4C')+
    theme_clean()
ggsave('./figure2/res9/HIS.pdf',width = 6,height = 7)
```
